```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Linear Regression

```python
# sample data
x = np.arange(10)
y = 5 * x + 10 

# Linear fit with np.polyfit -> (y = a * x + b)
a, b = np.polyfit(x, y, 1)

plt.plot(x, y, '.')
plt.plot(x, a * x + b, '-')
plt.title('Linear Regression')
plt.show()
```

## 2nd order Polynomial Regression

```python
# sample data
x = np.arange(20)
y = 2 * x ** 2 - 10 

# 2nd order fit with np.polyfit -> (y = a * x2 + b * x + c)
a, b, c = np.polyfit(x, y, 2)

plt.plot(x, y, '.', c='k')
plt.plot(x, a * x ** 2 + b * x + c, '--', c='r')
plt.title('Polynomial Regression')
plt.show()
```
